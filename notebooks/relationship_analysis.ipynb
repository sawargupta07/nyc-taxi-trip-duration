```python
# libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
base = "/content/drive/MyDrive/nyc-taxi-trip-duration"
df = pd.read_csv(f"{base}/data/train_clean.csv")
```

```python
# Correlation heatmap
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Passenger count analysis
sns.countplot(x='passenger_count', data=df)
plt.title("Passenger Count Frequency")
plt.show()
```

```python
# Trip duration vs Passenger count
sns.boxplot(x='passenger_count', y='trip_duration_min', data=df)
plt.title("Trip Duration vs Passenger Count")
plt.show()
```

```python
# Hourly trend
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['pickup_hour'] = df['pickup_datetime'].dt.hour
```

```python
sns.lineplot(x='pickup_hour', y='trip_duration_min', data=df)
plt.title("Average Trip Duration by Hour")
plt.show()
```
